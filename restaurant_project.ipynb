Cleaning data Column-wise

```python
#cleaning "rate" column

df['rate'].unique()
```

```python
#defining function to clean rate column

def cleanRate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
```

```python
df['rate'] = df['rate'].apply(cleanRate)
```

```python
df['rate'].unique()
```

Fill null value in cleaned column 'rate'

```python
df['rate'].fillna(df['rate'].mean(), inplace = True)
df['rate'].unique()
```

```python
df['rate'].isnull().sum()
```

```python
df.head()
```

```python
#location and listed_in(city) are same, so remove one
```

```python
df = df.drop(['listed_in(city)'], axis = 1)
```

```python
df.head()
```

```python
df['approx_cost(for two people)'].unique()
```

Removing "," from approx_cost(for two people) column

```python
def removeComma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',', '')
        return float(value)
    else:
        return float(value)    
```

```python
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].apply(removeComma)
```

```python
df['approx_cost(for two people)'].unique()
```

```python
df.head()
```

Cleaning Restaurant Type Column

```python
rest_types = df['rest_type'].value_counts(ascending  = False)
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]
rest_types_lessthan1000
```

Function: To make Restaurant Types < 1000 as others

```python
def rest_type_asOthers(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value
        
df['rest_type'] = df['rest_type'].apply(rest_type_asOthers)
df['rest_type'].value_counts()
```

Cleaning Location Column

```python
location = df['location'].value_counts(ascending  = False)
location
```

```python
location_lessthan300 = location[location<300]
location_lessthan300
```

Function: Converting Location column values < 300 as others

```python
def location_asOthers(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value

        
df['location'] = df['location'].apply(location_asOthers)
df['location'].value_counts()
```

Cleaning Cuisines column

```python
cuisines = df['cuisines'].value_counts(ascending  = False)
cuisines
```

```python
cuisines_lessthan100 = cuisines[cuisines<100]
cuisines_lessthan100
```

Function: Covert Cuisines column value < 100 as others

```python
def cuisines_asOthers(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value
        

df['cuisines'] = df['cuisines'].apply(cuisines_asOthers)
df['cuisines'].value_counts()
```

```python
df.head()
```

Renaming column names

```python
df.rename(columns = {'approx_cost(for two people)':'Cost2plates', 'listed_in(type)':'Type'}, inplace = True)
df.head()
```

# Wordcloud for dishes in each and every restaurant_type

```python
from wordcloud import WordCloud, STOPWORDS 
```

```python
df.isna().sum()
```

```python
df.dropna(axis='index',subset=['dish_liked'],inplace=True)
```

```python
df.isna().sum()
```

Word cloud for 1 Restaurant type

```python
data=df[df['rest_type']=='Quick Bites']
```

```python
data['dish_liked']
```

```python
stopwords=set(STOPWORDS)
```

```python
dishes=''
for word in data['dish_liked']:
    words=word.split()
    # Converts each token into lowercase 
    for i in range(len(words)): 
        words[i] = words[i].lower() 
    dishes=dishes+ " ".join(words)+" "
wordcloud = WordCloud(max_font_size=None, background_color='white', collocations=False,stopwords = stopwords,width=800, height=800).generate(dishes)
plt.imshow(wordcloud)
plt.axis("off")
```

Function: to generate word cloud for all restaurant types

```python
def generate_wordcloud(rest):
    
    plt.figure(figsize=(20,30))
    for i,restaurant in enumerate(rest):
        plt.subplot(3,3,i+1)
        dishes=''
        data=df[df['rest_type']==restaurant]
        for word in data['dish_liked']:
            words=word.split()
            # Converts each token into lowercase 
            for i in range(len(words)): 
                words[i] = words[i].lower() 
            dishes=dishes+ " ".join(words)+" "
        wordcloud = WordCloud(max_font_size=None, background_color='white', collocations=False,stopwords = stopwords,width=1500, height=1500).generate(dishes)
        plt.imshow(wordcloud)
        plt.title(restaurant)
        plt.axis("off")
```

```python
rest=df['rest_type'].value_counts()[:9].index
```

```python
stopwords = set(STOPWORDS) 
generate_wordcloud(rest)
```

Reveiew analysis using Word cloud on all types of restaurant

```python
def review_analysis(restaurant):
    dataset=df[df['rest_type']==restaurant]
    total_review=' '
    for review in dataset['reviews_list']:
        review=review.lower()
        review=re.sub('[^a-zA-Z]', ' ',review)
        review=re.sub('rated', ' ',review)
        review=re.sub('x',' ',review)
        review=re.sub(' +',' ',review)
        total_review=total_review + str(review)
    wordcloud = WordCloud(width = 800, height = 800, 
            background_color ='white', 
            stopwords = set(STOPWORDS), 
            min_font_size = 10).generate(total_review) 
    # plot the WordCloud image                        
    #plt.figure(figsize = (8, 8)) 
    plt.imshow(wordcloud) 
    plt.axis("off") 
    plt.title(restaurant)
```

```python
plt.figure(figsize=(20,30))
for i,restaurant_types in enumerate(rest):
  plt.subplot(3,3,i+1)
  review_analysis(restaurant_types)
```
